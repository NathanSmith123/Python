```python
import pandas as pd
import numpy as np

csvfile = 'https://www.dropbox.com/s/jgtk2x9l1kh8oz4/Travel%20Pony%20Facebook.csv?dl=1'

FBPony = pd.read_csv(csvfile)
```

```python
#Calculates the cost per impressions and then adds that data as a column
Cost_Per_Impression = (FBPony.loc[:,'Amount Spent (USD)'])/(FBPony.loc[:,'Impressions'])
FBPony['Cost Per Impression'] = Cost_Per_Impression
```

```python
import arrow

#initializes a place to store the day of the week 
dayofweek = []

#loop that iterates through FBPony data and gets the start date value
for row in range(len(FBPony)):
    date = arrow.get(FBPony['Start Date'][row], 'M/D/YY')
    dayofweek.append(date.isoweekday())
    
#takes the iterated date data from the above for loop and stores it
FBPony['Day'] = dayofweek

#groups the days by each day
group = FBPony.groupby('Day')

#calculates the mean for each day group
group.aggregate(np.mean)
#Friday is the most expensive day. Thursday and Sunday are also very expensive
#Saturday is the cheapest day.
```

```python
#calculate the correlation of amount spent compared to reach, freq, unique clicks, and page likes
ASandR = np.corrcoef(FBPony['Amount Spent (USD)'],FBPony['Reach'])
ASandFreq = np.corrcoef(FBPony['Amount Spent (USD)'],FBPony['Frequency'])
ASandUC = np.corrcoef(FBPony['Amount Spent (USD)'],FBPony['Unique Clicks'])
ASandPl = np.corrcoef(FBPony['Amount Spent (USD)'],FBPony['Page Likes'])
print('ASandR')
print(ASandR)
###
print('ASandFreq')
print(ASandFreq)
###
print('ASandUC')
print(ASandUC)
###
print('ASandPl')
print(ASandPl)
```

```python
import statsmodels.api as sm
```
